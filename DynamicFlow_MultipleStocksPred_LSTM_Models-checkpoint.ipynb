```python
def get_data( stock_symbol,from_date,to_date):
    list_of_stocks = []
    for i in range(0,len(stock_symbol)):
        data = instruments[(instruments['tradingsymbol'] == stock_symbol[i]) & (instruments['instrument_type'] == 'EQ')]
        data = data.head(1)
    #from_date = '2017-03-01'
    #to_date = '2021-12-22'
        ttoken = data['instrument_token'].item()
        dt_stock = pd.DataFrame(kite.historical_data(ttoken, from_date, to_date,'day',0,oi=1))
        dt_stock['Symbol'] = stock_symbol[i]
        list_of_stocks.append(dt_stock)
    return list_of_stocks
```

```python
def preprocessing_data(list_df_stocks, stock_symbol):
    list_df_preprocessed_stocks = []
    for i in range(0,len(stock_symbol)):
        dt_stock = list_df_stocks[stock_symbol.index(stock_symbol[i])]
        dt_stock["returns_numeric"] = dt_stock.close - dt_stock.close.shift(1)
        dt_stock["returns_percentage"] = (dt_stock.returns_numeric/ dt_stock.close.shift(1))*100
        dt_stock["volume_change_numeric"] = dt_stock.volume - dt_stock.volume.shift(1)
        dt_stock["volume_change_percentage"] = (dt_stock.volume_change_numeric / dt_stock.volume.shift(1))*100
        dt_stock["2D_SMA"]= dt_stock['close'].rolling(2).mean()
        dt_stock["3D_SMA"]= dt_stock['close'].rolling(3).mean()
        dt_stock["7D_SMA"]= dt_stock['close'].rolling(7).mean()
        dt_stock["21D_SMA"]= dt_stock['close'].rolling(21).mean()
        dt_stock["90D_SMA"]= dt_stock['close'].rolling(90).mean()
        dt_stock["Expanding_Mean"] = dt_stock['close'].expanding().mean()
        dt_stock["3D_Var"] = dt_stock['close'].rolling(3).var()
        dt_stock["7D_Var"] = dt_stock['close'].rolling(7).var()
        dt_stock["21D_Var"] = dt_stock['close'].rolling(21).var()
        dt_stock["90D_Var"] = dt_stock['close'].rolling(90).var()
        dt_stock["Expanding_Var"] = dt_stock['close'].expanding().var()
        dt_stock["3D_EMA"]= dt_stock['close'].ewm(span=3,adjust=False).mean()
        dt_stock['7D_EMA'] = dt_stock.close.ewm(span=7,adjust=False).mean()
        dt_stock['21D_EMA'] = dt_stock.close.ewm(span=21,adjust=False).mean()
        dt_stock['90D_EMA'] = dt_stock.close.ewm(span=90,adjust=False).mean()
        dt_stock["C_O_Diff"] = dt_stock["close"] - dt_stock["open"]
        #print(type(dt_stock),i)
        dt_stock = dt_stock.dropna()
        list_df_preprocessed_stocks.append(dt_stock)
        #print(type(list_df_preprocessed_stocks),i)

        
    return list_df_preprocessed_stocks
```

```python
def select_features(list_df_preprocessed_stocks, stock_symbol, list_features):

    list_df_preprocessed_features_stocks = []
    for i in range(0,len(stock_symbol)):
        print(i)
        dt_stock = list_df_preprocessed_stocks[i]
        dt_stock = dt_stock.dropna()
        no_of_features = len(list_features)
        data = dt_stock.sort_index(ascending=True, axis=0)
        new_data = pd.DataFrame(index=range(0,len(dt_stock)),columns=list_features)
        new_data['date'] = data['date']
        #new_data['class'] = data['class']
        new_data['close'] = data['close']
        new_data['volume'] = data['volume']
        #new_data['volume_change_percentage'][i] = data['volume_change_percentage'][i]
        new_data['3D_SMA'] = data['7D_SMA']
        new_data['7D_SMA'] = data['7D_SMA']
        new_data['3D_EMA'] = data['3D_EMA']
        new_data['7D_EMA'] = data['7D_EMA']
        new_data['3D_Var'] = data['3D_Var']
        new_data['7D_Var'] = data['7D_Var']
        new_data['C_O_Diff'] = data['C_O_Diff']
        new_data.index = new_data.date
        new_data.drop('date', axis=1, inplace=True)
        new_data = new_data.dropna()
        print("data shape",new_data.shape)
        #dataset = new_data.values
        #list.append(dataset)
        #print(dataset.shape)
        list_df_preprocessed_features_stocks.append(new_data)
    return list_df_preprocessed_features_stocks
```

```python
def scaling_data(list_df_preprocessed_stocks):
    list_dfs = []
    global list_scale
    list_scale = []
    for i in range(0,len(list_df_preprocessed_stocks)):
        scaler = MinMaxScaler(feature_range=(0, 1))
        scaled_data = scaler.fit_transform(list_df_preprocessed_stocks[i])
        #print("shape",scaled_data.shape)
        list_dfs.append(scaled_data)
        list_scale.append(scaler)
    print("Scaling done")
    return list_dfs
```

```python
# create and fit the LSTM network
def build_LSTM_model(x_train,y_train):
    #list_models = []
    for i in range(0,len(stock_symbol)):
        print("##############################################################################################")
        print("@@@@@@@@@@@@@@@@@@@@@@Starting to Build Model for",stock_symbol[i],"@@@@@@@@@@@@@@@@@@@@@@@@@")
        model = Sequential()
        model.add(LSTM(units=64, return_sequences=True, input_shape=(x_train.shape[1],1)))
        model.add(LSTM(units=32))
        model.add(Dense(1))
        model.compile(loss='mean_squared_error', optimizer='adam')
        print("shape of x_train",x_train.shape)
        print("shape of y_train",y_train.shape)
        model.fit(tf.convert_to_tensor(x_train), tf.convert_to_tensor(y_train), epochs=20, batch_size=50, verbose=2)
        #list_models.append(model)
        print("@@@@@@@@@@@@@@@@@@@@@@Built Model for",stock_symbol[i],"@@@@@@@@@@@@@@@@@@@@@@@@@")
    return model
```

```python
def predict_from_model(list_models, list_train_test_x_y_timesteps,stock_symbol, no_of_features):
    list_y_pred = []
    for i in range(0,len(list_models)):
        y_pred = list_models[i].predict(list_train_test_x_y_timesteps[i][0][0])
        #print(y_pred.shape)
        #print(y_pred.reshape(y_pred.shape[0],1))
        #y_pred = y_pred.reshape(y_pred.shape[0],1)
        df_cp = pd.DataFrame(y_pred)
        for j in range(1,(no_of_features+1)):
            df_cp[j] = df_cp[0]
        #print(df_cp.shape)
        s = list_scale[i]
        df_cp = s.inverse_transform(df_cp)
        df_cp = df_cp[0:len(df_cp),0]
        list_y_pred.append(df_cp)
    return list_y_pred
    
    
    
```

```python
def get_y_actual(list_train_test_x_y_timesteps):
    list_y_actual = []
    for i in range(0,len(list_train_test_x_y_timesteps)):
        y_train = list_train_test_x_y_timesteps[i][0][1]
        a = y_train.reshape(y_train.shape[0],1)
        a.shape
        a = pd.DataFrame(a)
        for j in range(1,(no_of_features+1)):
            a[j] = a[0]
        #print(a.shape)
        s = list_scale[i]
        a = s.inverse_transform(a)
        a = a[0:len(a),0]
        list_y_actual.append(a)
    return list_y_actual
        
```

```python
def error_metrics(list_y_pred, list_y_actual):
    for i in range(0,len(list_y_actual)):
        print(list_y_pred[i].shape)
        rms=np.sqrt(np.mean(np.power((list_y_pred[i]-list_y_actual[i]),2)))
        print("rms",rms)
        mae=np.mean(np.absolute(list_y_pred[i]-list_y_actual[i]))
        print("mae",mae)
```
